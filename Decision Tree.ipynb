```python
#first we import the library and  load the data  
import pandas as pd
dataset=pd.read_csv('kidney_disease.csv')
dataset.head()
```

```python
dataset.shape
```

```python
#notice that we have missing values and nominal values
dataset.info()
```

```python
#Handling Missing float Values by replace them with the median

from sklearn.impute import SimpleImputer 
imputer = SimpleImputer(strategy="mean")
dataset.age=imputer.fit_transform(dataset['age'].values.reshape(-1,1))
dataset.bp =imputer.fit_transform(dataset['bp'].values.reshape(-1,1))
dataset.sg=imputer.fit_transform(dataset['sg'].values.reshape(-1,1))
dataset.al =imputer.fit_transform(dataset['al'].values.reshape(-1,1))
dataset.su=imputer.fit_transform(dataset['su'].values.reshape(-1,1))
dataset.bgr=imputer.fit_transform(dataset['bgr'].values.reshape(-1,1))
dataset.bu=imputer.fit_transform(dataset['bu'].values.reshape(-1,1))
dataset.bgr=imputer.fit_transform(dataset['bgr'].values.reshape(-1,1))
dataset.sc=imputer.fit_transform(dataset['sc'].values.reshape(-1,1))
dataset.sod=imputer.fit_transform(dataset['sod'].values.reshape(-1,1))
dataset.pot=imputer.fit_transform(dataset['pot'].values.reshape(-1,1))
dataset.hemo=imputer.fit_transform(dataset['hemo'].values.reshape(-1,1))
dataset.pcv=imputer.fit_transform(dataset['bgr'].values.reshape(-1,1))

dataset.info()
```

```python
#Handling Missing object Values by replace them with the most frequant

imputer_obj = SimpleImputer(strategy="most_frequent")
dataset.rbc=imputer_obj.fit_transform(dataset['rbc'].values.reshape(-1,1))
dataset.pc=imputer_obj.fit_transform(dataset['pc'].values.reshape(-1,1))
dataset.pcc=imputer_obj.fit_transform(dataset['pcc'].values.reshape(-1,1))
dataset.ba=imputer_obj.fit_transform(dataset['ba'].values.reshape(-1,1))
dataset.wc=imputer_obj.fit_transform(dataset['wc'].values.reshape(-1,1))
dataset.rc=imputer_obj.fit_transform(dataset['rc'].values.reshape(-1,1))
dataset.htn=imputer_obj.fit_transform(dataset['htn'].values.reshape(-1,1))
dataset.dm=imputer_obj.fit_transform(dataset['dm'].values.reshape(-1,1))
dataset.cad=imputer_obj.fit_transform(dataset['cad'].values.reshape(-1,1))

dataset.info()
```

```python
dataset
```

```python
#We transform our features and the target variable to numeric data
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()
dataset.rbc=label_encoder.fit_transform(dataset[['rbc']])
dataset.pc=label_encoder.fit_transform(dataset[['pc']])
dataset.pcc=label_encoder.fit_transform(dataset[['pcc']])
dataset.ba=label_encoder.fit_transform(dataset[['ba']])
dataset.rc=label_encoder.fit_transform(dataset[['rc']])
dataset.wc=label_encoder.fit_transform(dataset[['wc']])
dataset.htn=label_encoder.fit_transform(dataset[['htn']])
dataset.dm=label_encoder.fit_transform(dataset[['dm']])
dataset.cad=label_encoder.fit_transform(dataset[['cad']])
dataset.appet=label_encoder.fit_transform(dataset[['appet']])
dataset.pe=label_encoder.fit_transform(dataset[['pe']])
dataset.ane=label_encoder.fit_transform(dataset[['ane']])
dataset.classification=label_encoder.fit_transform(dataset[['classification']])
```

```python
dataset
```

```python
#spliting the Data set to apply the model into train and test
from sklearn.model_selection import train_test_split
#split the features and target
x=dataset.iloc[:,:-1]
y=dataset.iloc[:,-1]
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3)
```

```python
x
```

```python
y
```

```python
#now we select the most 15% important features from the dataset 
from sklearn.feature_selection import SelectPercentile   , chi2 
print(x.shape)
select = SelectPercentile(score_func = chi2 , percentile = 15)
new_x= select.fit_transform(x,y)
print(new_x.shape)
```

```python
#now tha data is ready to apply the  model

from sklearn.tree import DecisionTreeClassifier
DecisionTreeClassifierModel = DecisionTreeClassifier(criterion='entropy',max_depth=6) 
DecisionTreeClassifierModel.fit(x_train, y_train)
```

```python
new_x
```

```python
#now we predict the test data
y_pred= DecisionTreeClassifierModel.predict(x_test)
```

```python
#evaluate the  model 
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report

print('the accuracy of the third modle :',accuracy_score(y_test,y_pred))
print('the confusion matrix',confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```
